### doing some visualisation and feature understanding

(univariate analysis)
- plotting feature distributitons
- histogram
- KDE 
- Boxplot

```python
#how many are there for each country
wheat['Country'].value_counts() .plot(kind= 'bar') #all datapoints
```

```python

for country in wheat['Country'].unique():
    subset = wheat[wheat['Country'] == country]
    
    # First define yield_values
    yield_values = subset['Grain yield (tons ha-1)']
    
    # Try to convert to numeric
    numeric_yields = pd.to_numeric(yield_values, errors='coerce')
    
    # Filter outliers
    reasonable_max = 30  # removing obvious outliers
    numeric_yields = numeric_yields[numeric_yields <= reasonable_max]
    clean_yields = numeric_yields.dropna()
    
    # Statistical test first (before plots)
    if len(clean_yields) >= 3:
        shapiro_test = stats.shapiro(clean_yields)
        is_normal = 'Yes' if shapiro_test[1] > 0.05 else 'No'
        shapiro_stat = f"{shapiro_test[0]:.4f}"
        shapiro_p = f"{shapiro_test[1]:.4f}"
    else:
        is_normal = 'N/A'
        shapiro_stat = 'N/A'
        shapiro_p = 'N/A'
    
    # Print info with Shapiro results
    print(f"Country: {country}")
    print(f"  Number of rows: {len(subset)}")
    print(f"  Data type: {yield_values.dtype}")
    print(f"  Non-NA values: {yield_values.notna().sum()}")
    print(f"  Shapiro-Wilk statistic: {shapiro_stat}")
    print(f"  Shapiro-Wilk p-value: {shapiro_p}")
    print(f"  Normal distribution? {is_normal}")
    
    # Plot only if enough data
    if len(clean_yields) > 2:
        try:
            # Create 3 subplots side by side
            fig, (hist_ax, kde_ax, qq_ax) = plt.subplots(1, 3, figsize=(18, 5))
            
            # Histogram plot
            clean_yields.hist(ax=hist_ax, bins=20)
            hist_ax.set_title(f'{country} - Histogram')
            hist_ax.set_xlabel('Grain yield (tons ha-1)')
            
            # KDE plot
            clean_yields.plot.kde(ax=kde_ax, label='Actual')
            x = np.linspace(clean_yields.min(), clean_yields.max(), 100)
            kde_ax.plot(x, stats.norm.pdf(x, clean_yields.mean(), clean_yields.std()), 
                    'r--', label='Normal')
            kde_ax.set_title(f'{country} - Yield Distribution')
            kde_ax.set_xlabel('Grain yield (tons ha-1)')
            kde_ax.legend()
            
            # QQ plot
            stats.probplot(clean_yields, dist="norm", plot=qq_ax)
            qq_ax.set_title(f'{country} - QQ Plot')
            
            plt.tight_layout()
            plt.show()
        except Exception as e:
            print(f"  Error plotting {country}: {str(e)}")
    else:
        print(f"  Skipping plots - insufficient data")
    
    print('-' * 50)
```
